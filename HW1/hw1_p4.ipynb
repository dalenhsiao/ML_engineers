Imports and provided functions:

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2 + 3*x + 6*np.sin(x)

def plotfx():
    # Sample function
    xs = np.linspace(-12,10,100)
    ys = f(xs)
    # Plot function
    plt.plot(xs,ys,'r-')
    plt.xlabel('$x$')
    plt.ylabel('$f(x)$')
    plt.show()

# Visualize the function
plotfx()
```

### First define the function $\texttt{fgrad}(x)$

```python
# Your fgrad(x) function goes here
def fgrad(x): 
    f = 2*x+6*np.cos(x)+3
    return f
```

### Fill in the following code with the gradient descent update rule
For reference, your 10th iteration should have $x = -1.554$ and $f(x) = -8.246$

```python
iter = 10
eta = 0.15
x = 8
x_store = [x]

def plot_traversal():
    # Sample function
    xs = np.linspace(-12,10,100)
    ys = f(xs)
    # Plot function
    plt.plot(xs,ys,'r-')
    plt.xlabel('$x$')
    plt.ylabel('$f(x)$')

for i in range(iter):
    # YOUR GRADIENT DESCENT CODE GOES HERE
    x -= eta*fgrad(x)
    x_store.append(x)
    print('Iteration %d, x = %.3f, f(x) = %.3f' %(i+1, x, f(x)))
    print("Gradient: %.3f" %fgrad(x))

y_store = [f(x) for x in x_store]

plot_traversal()
plt.plot(x_store, y_store, marker="o")
plt.legend(["f(x)", "gradient decent"])
plt.show()

    

```
